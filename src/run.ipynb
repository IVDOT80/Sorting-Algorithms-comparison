```python
from bubble_sort import bubble_sort
from insertion_sort import insertion_sort
from merge_sort import merge_sort
from quick_sort import quick_sort
from selection_sort import selection_sort
from utils import measure_space, measure_time
```

```python
sample_size = [10, 100, 1000, 10000]
algorithms_names = {'bubble_sort': bubble_sort, 'insertion_sort': insertion_sort,
                    'merge_sort': merge_sort, 'quick_sort': quick_sort,
                    'selection_sort': selection_sort}
```

```python
algorithms = dict()
for name in algorithms_names.keys():
    algorithms[name] = {'time': [], 'complexity': []}
```

```python
algorithms
```

```python
for size in sample_size:
    list_unsort = [int(random.uniform(-1000, 1000)) for _ in range(size)]

    for algorithm in algorithms_names:
        algorithms[algorithm]['time'] = measure_time(algorithms_names[algorithm], list_unsort)
        algorithms[algorithm]['complexity'] = measure_space(algorithms_names[algorithm], list_unsort)

    
```
